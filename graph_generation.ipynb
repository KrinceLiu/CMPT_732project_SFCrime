```python
import folium
import pandas as pd
```

```python
df_crime_2016 = pd.read_csv('./data/Police_Department_Incidents_-_Previous_Year__2016_.csv')
```

```python
df_crime_2016.head()
```

```python
RECORDS_SHOW = 500
SF_LOCATION = (37.76, -122.45)
map = folium.Map(
    location=SF_LOCATION, 
    zoom_start=12,
    title='SF Crime map')
for each in df_crime_2016[0:RECORDS_SHOW].iterrows():
    folium.CircleMarker(
        location = [each[1]['Y'],each[1]['X']], 
        clustered_marker = True,
        color='crimson',
        fill=True,
        fill_color='crimson'
    ).add_to(map)
  
display(map)
```

```python
import matplotlib
import seaborn as sns
import matplotlib.pyplot as plt

week_x = ['Mon', 'Tue', 'Wed', 'Thu', 'Fir', 'Sat','Sun']
df_weekday = pd.to_datetime(df_crime_16years_2['Date']).dt.dayofweek.to_frame()
df_crime_16years_2['weekday'] = df_weekday
df_weekday['count'] = 1
df_count = df_weekday.groupby('Date').sum()
df_count['weekday'] = week_x

g = sns.catplot(x='weekday', y='count', kind="bar", data=df_count)
g.fig.set_figwidth(7)
g.fig.set_figheight(3)
```

```python
df_crime_16years_2['month'] = pd.DatetimeIndex(pd.to_datetime(df_crime_16years_2['Date'])).month
df_count_month = df_crime_16years_2['month'].to_frame()
df_count_month['count'] = 1
df_count_month = df_count_month.groupby('month').sum()
df_count_month.reset_index(inplace=True)

g = sns.catplot(x='month', y='count', kind="bar", data=df_count_month)
g.fig.set_figwidth(10)
g.fig.set_figheight(3)
```

```python
df_crime_16years_2['year'] = pd.DatetimeIndex(pd.to_datetime(df_crime_16years_2['Date'])).year
df_count_year = df_crime_16years_2['year'].to_frame()
df_count_year['count'] = 1
df_count_year = df_count_year.groupby('year').sum()
df_count_year.reset_index(inplace=True)



g = sns.catplot(x='year', y='count', kind="bar", data=df_count_year)
g.fig.set_figwidth(10)
g.fig.set_figheight(3)
```
